```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

def update_point(ind, positions_of_lidar, points):
    for point in points:
        X, Y, Z = positions_of_lidar['positions_x'][0][ind], positions_of_lidar['positions_y'][0][ind], positions_of_lidar['positions_z'][0][ind]
        ret = np.vstack((Y, X))
        point.set_data(ret.T)
        point.set_3d_properties(Z.T)
    return points
    
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
points = [ax.plot([], [], [], 'g.')[0] for _ in range(len(positions_of_lidar['positions_x'][0]))]
# lines = [ax.plot([], [], [])[0]]
ax.set(xlim3d=(15, -15), xlabel='Y')
ax.set(ylim3d=(-60, 60), ylabel='X')
ax.set(zlim3d=(-5, 5), zlabel='Z')

ani = animation.FuncAnimation(
    fig, update_point, frames=594, interval=20, fargs=(positions_of_lidar, points))

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Fixing random state for reproducibility
np.random.seed(19680801)


def random_walk(num_steps, max_step=0.05):
    """Return a 3D random walk as (num_steps, 3) array."""
    start_pos = np.random.random(3)
    steps = np.random.uniform(-max_step, max_step, size=(num_steps, 3))
    walk = start_pos + np.cumsum(steps, axis=0)
    return walk


def update_lines(num, walks, lines):
    for line, walk in zip(lines, walks):
        # NOTE: there is no .set_data() for 3 dim data...
        line.set_data(walk[:num, :2].T)
        line.set_3d_properties(walk[:num, 2])
        print(walk.shape)
    return lines


# Data: 40 random walks as (num_steps, 3) arrays
num_steps = 30 # 30 스텝으로 보인다.
walks = [random_walk(num_steps) for index in range(40)] # 점 40개
walks = np.array(walks)
print(walks.shape)
print(len(walks))

# Attaching 3D axis to the figure
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

# Create lines initially without data
lines = [ax.plot([], [], [])[0] for _ in walks]

# Setting the axes properties
ax.set(xlim3d=(0, 1), xlabel='X')
ax.set(ylim3d=(0, 1), ylabel='Y')
ax.set(zlim3d=(0, 1), zlabel='Z')

# Creating the Animation object
# ani = animation.FuncAnimation(
#     fig, update_lines, num_steps, fargs=(walks, lines), interval=100)

# plt.show()
```
